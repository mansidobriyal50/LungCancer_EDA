```python
#Gender Distribution
plt.figure(figsize=(6,5))
df.GENDER.value_counts().plot.pie(autopct='%1.1f%%',colors=['lightcoral','lightgreen'],startangle=90,shadow=True)
```

Gender Distribution:
The dataset appears to have an equal distribution of male and female patient. This balanced distribution helps in avoiding gender bias in the analysis of lung cancer risk factors.

```python
# Set the style for the plots
sns.set(style="whitegrid")
```

```python
# 2. Age Distribution
plt.figure(figsize=(8, 6))
sns.histplot(df['AGE'], bins=20, kde=True, color='blue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

The age distribution is approximately normal, with most participants clustered around the mean age of 55. This suggests that the study primarily focuses on middle-aged and older individuals, which is relevant since lung cancer risk increases with age.

```python
# 3. Smoking and Lung Cancer
plt.figure(figsize=(8, 6))
sns.countplot(x='SMOKING', hue='LUNG_CANCER', data=df, palette='Set1')
plt.title('Smoking vs Lung Cancer')
plt.xlabel('Smoking')
plt.ylabel('Count')
plt.legend(title='Lung Cancer')
plt.show()
```

```python
#Smoking vs. Lung Cancer:
#The count plot shows a clear relationship between smoking and lung cancer. Participants who smoke are significantly more likely to have lung cancer compared to non-smokers. This reinforces the well-known link between smoking and lung cancer risk.
```

```python
df.columns
```

```python
corr=df.corr(numeric_only=True)
plt.figure(figsize=(13,8))
sns.heatmap(corr,annot=True,cmap='coolwarm')
plt.title('Correlation Matrix')
```

Correlation Matrix:
The heatmap of the correlation matrix shows that smoking has a strong positive correlation with lung cancer. Additionally, there are other features like anxiety and chronic disease that also show correlations with lung cancer, though to a lesser extent.
Features like gender and age do not have strong correlations with lung cancer in this dataset, suggesting that these factors alone may not be primary indicators without considering other variables.
